```python
fruits = ['pineapples', 'bananas', 'lemons', 'apples', 'pears']
quantity = [100, 150, 85, 420, 200]
```

# Задача 2

#### Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
#### [100, 125, -90, 345, 655, -1, 0, 200]
#### Используйте list comprehensions

```python
list_pay = [100, 125, -90, 345, 655, -1, 0, 200]
```

```python
def new_list_pay(pay):
    return pay if pay > 0 else 0

new_pay = [new_list_pay(i) for i in list_pay]
new_pay
```

```python
sum(new_pay)
```

## Используйте list comprehensions

```python
date = ['2021-09-14', '2021-12-15', '2021-09-08',
        '2021-12-05', '2021-10-09', '2021-09-30',
        '2021-12-22', '2021-11-29', '2021-12-24',
        '2021-11-26', '2021-10-27', '2021-12-18',
        '2021-11-09', '2021-11-23', '2021-09-27',
        '2021-10-02', '2021-12-27', '2021-09-20',
        '2021-12-13', '2021-11-01', '2021-11-09',
        '2021-12-06', '2021-12-08', '2021-10-09',
        '2021-10-31', '2021-09-30', '2021-11-09',
        '2021-12-13', '2021-10-26', '2021-12-09']

payment = [1270, 8413, 9028, 3703, 5739, 4095,
           295, 4944, 5723, 3701, 4471, 651,
           7037, 4274, 6275, 4988, 6930, 2971,
           6592, 2004, 2822, 519, 3406, 2732,
           5015, 2008, 316, 6333, 5700, 2887]
```

```python
total_list = {p: date[d] for d, p in enumerate(payment)}
```

```python
for d, p in zip(date, payment):
    total_list[d].add(p)
```

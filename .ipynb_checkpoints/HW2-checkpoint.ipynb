## Задача 5
_Ответьте на несколько вопросов_

* 5.1 Выберите дом клиенту.
Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?

```python
df[(df['waterfront'] > 0) & (df['bathrooms'] >= 3) & (df['sqft_basement'] > 0)]
```

_Итого: 41 вариант_

* 5.2 Выберите дом клиенту.
Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

```python
range_price = df[(df['yr_built'] >= 1980) & ((df['waterfront'] > 0) | (df['view'] > 0))]
range_price
```

```python
range_price.describe()
```

```python
range_price['price'].min()
```

```python
range_price['price'].max()
```

_Дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года можно приобрести в ценовоь дипазоне от 245 000 до 6 885 000._

* 5.3 Выберите дом клиенту
Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?

```python
mean_grade = df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] <= 150000)]
mean_grade
```

```python
mean_grade.describe()
```

```python
mean_grade['grade'].mean()
```

_Оценка по состоянию у таких домов в среднем равна 7_
